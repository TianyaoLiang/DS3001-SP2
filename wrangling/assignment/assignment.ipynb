**Q3.** Open the "tidy_data.pdf" document in the repo, which is a paper called Tidy Data by Hadley Wickham.

  1. Read the abstract. What is this paper about?
  
  The paper is about teaching a universal framework on how data should me cleaned and presented

  2. Read the introduction. What is the "tidy data standard" intended to accomplish?

  The tidy data standard is intended to be designed to facilitate initial exploration and analysis of the data, and to simplify the development of data analysis tools that work well together.

  3. Read the intro to section 2. What does this sentence mean: "Like families, tidy datasets are all alike but every messy dataset is messy in its own way." What does this sentence mean: "For a given dataset, its usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general." 

  To me, the first sentence mean that datasets are related like family members but each member is unique so you need something to tie them together. The second sentences means that it is easy to see the data, but it is hard to interpret them.

  4. Read Section 2.2. How does Wickham define values, variables, and observations?

  Values are either quantitative or qualitative and they belong to varaibles and observations. A variable contains values that measure the same attribute across units. Observations are values measured in the same unit like person or race across attributes. 

  5. How is "Tidy Data" defined in section 2.3?

  Each variable forms a column, each observation forms a row, and each type of observational unit forms a table.

  6. Read the intro to Section 3 and Section 3.1. What are the 5 most common problems with messy datasets? Why are the data in Table 4 messy? What is "melting" a dataset?

  Top 5 most common problems are column headers are values, not variable names, multiple variables are stored in one column, variables are stored in both rows and columns, multiple types of observational units are stored in the same table, and a single observational unit is stored in multiple tables.
  Table 4 is messy because it is a tabular data table, where variables form both the rows and columns, and column headers are values, not variable names.
  Melting a dataset means turning columns into rows.

  7. Why, specifically, is table 11 messy but table 12 tidy and "molten"?

  Table 11 is messy because the element column is not a variable and it stores the names of variables. Table 12 is tidy because it is transformed/molten with columns that contains variable names and element.

  8. Read Section 6. What is the "chicken-and-egg" problem with focusing on tidy data? What does Wickham hope happens in the future with further work on the subject of data wrangling?'

  The "chicken-and-egg" problem with focusing on tidy data is the concept that tidy data is only useful if tools support it and tools are only developed if tidy data is common. Therefore, it makes progress slow because data formats and tools depend on each other. Wichham hopes that in the future researchers can build better tools for working with data and develope new frameworks to improve data wrangling and cleaning.
